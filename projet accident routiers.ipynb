```python
# Détection par IQR 
def detect_outliers_iqr(df, column): 
    Q1 = df[column].quantile(0.25) 
    Q3 = df[column].quantile(0.75) 
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR 
    outliers = df[(df[column]< lower_bound) | (df[column] >upper_bound)] 
    return outliers, lower_bound, upper_bound 
```

```python
# Utilisation 
outliers, lower, upper = detect_outliers_iqr(cleaned2024, 'vma') 
print(f"Nombre d'outliers: {len(outliers)}") 
print(f"Limites: [{lower:.2f}, {upper:.2f}]")
```

```python
from sklearn.ensemble import RandomForestRegressor 
# Pour chaque colonne avec des valeurs manquantes 
for col in df_test.columns[df_test.isnull().any()]: 
    # Données d'entraînement (lignes sans NaN pour cette colonne) 
    train = df_test[df_test[col].notna()] 
    test = df_test[df_test[col].isna()] 
    # Features et target 
    features = [c for c in df_test.columns if c != col] 
    X_train = train[features] 
    y_train = train[col] 
    X_test = test[features] 
    # Entraîner le modèle et prédire 
    model = RandomForestRegressor(n_estimators=100, random_state=42) 
    model.fit(X_train, y_train) 
    df_test.loc[df_test[col].isna(), col] = model.predict(X_test)
```

## EDA et features

```python
df_cleaned = df_gen
```

```python
# transformer les colonnes secu1, secu2, secu3 et secu en une seule colonne qui contient l'utilisation d'un équipement de sécurité
```

```python
def nouvelle_valeur(valeur):
    protege = [1, 2, 3, 4, 5, 6, 7, 11, 21, 31, 41, 91]
    non_protege = [0, 12, 22, 32, 42, 92]
    if valeur.isin(protege).any():
            return 1
    elif valeur.isin(non_protege).any():
            return 2
    else :
            return 3

df_cleaned['securite'] = df_cleaned[['secu1', 'secu2', 'secu3', 'secu']].apply(nouvelle_valeur, axis = 1)
df_cleaned['securite'].value_counts()
```

```python
# créer une colonne bool sur la colonne locp : 0 pour aucun pieton et 1 si pieton présent
```

```python
def valeur_piet(valeur):
    chausse = [1, 2]
    passage_pieton = [3, 4]    
    if valeur in chausse :
            return 1
    elif valeur in passage_pieton :
            return 2
    else :
            return 3

df_cleaned['loc_piet'] = df_cleaned['locp'].apply(valeur_piet)
df_cleaned
```

```python
# créer une colonne bool sur la colonne grav : 0 pour mort et 1 pour vivant
```

```python
def valeur_grav(valeur):
    indemne = [1]
    mort = [2]    
    if valeur in indemne :
            return 1
    elif valeur in mort :
            return 2
    else :
            return 3

df_cleaned['gravite'] = df_cleaned['grav'].apply(valeur_grav)
df_cleaned
```

```python
# gestion des jours
```

```python
def valeur_jour(valeur):    
    if valeur <= 15 :
            return 1
    else :
            return 2

df_cleaned['mi_mois'] = df_cleaned['jour'].apply(valeur_jour)
df_cleaned
```

```python
# suppression des colonnes qui ne sont plus nécéssaires
df_final = df_cleaned.drop(columns = 
                           ['secu1',
                            'secu2', 
                            'secu3', 
                            'secu', 
                            'jour', 
                            'hrmn', 
                            'locp', 
                            'grav', 
                            'lat', 
                            'long',
                            'adr',
                            'voie',
                            'v1', 
                            'v2', 
                            'lartpc', 
                            'larrout', 
                            'gps', 
                            'pr',
                            'dep',
                            'com',
                            'id_vehicule',
                            'id_usager',
                            'Num_Acc',
                            'num_veh',
                            'env1',
                            'pr1'])
df_final.dtypes
```

## Modélisation

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import (confusion_matrix, classification_report)
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df_final.drop(columns=['gravite'])
y = df_final['gravite']
df_model = pd.concat([X, y], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle = True, stratify = y)
```

```python
target_names = ['indemne', 'mort', 'blessé hospitalisé', 'blessé léger']
```

### Régression Logistique

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels = target_names, yticklabels = target_names,
            cbar_kws={'label': 'Nombre de prédictions'},
            linewidths=2, linecolor='white')
plt.title('Matrice de Confusion - Classification\n', fontsize=16, fontweight='bold')
plt.ylabel('Vraie classe', fontsize=12)
plt.xlabel('Classe prédite', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
print(classification_report(y_test, y_pred, target_names= target_names))
```

### KNeighbors Classifier
